```python
percent = 0
X = []
Y = []
imageSize = 300

for pokemon in pokemonList:
  print("Now Loading " + str(round(percent / len(pokemonList) * 100, 2)) + "% Pokemon: " + pokemon)
  path = os.path.join(dir, pokemon)
  if pokemon == "Nidoqueen":
    break
  try:
    for image in os.listdir(path):
      try:
        imageStore = imread(os.path.join(path, image))
      except:
        continue
      imageStore = resize(imageStore, (imageSize, imageSize, 3)) #resizeImage
      X.append(imageStore.flatten())
      Y.append(percent)
    percent += 1
  except:
    percent += 1
    continue
  

x = np.array(X)
y = np.array(Y)

X_train, X_val, y_train, y_val = train_test_split(X,y, test_size = 0.2, train_size = 0.8)
```

```python
from sklearn.multiclass import OneVsRestClassifier
#BEST MODEL
n = 10
svc = SVC(kernel='linear',gamma='auto')
svc.fit(X_train, y_train)
valPredict = svc.predict(X_val)
#Print Accuracy
print("Val Set Accuracy:", str(round(accuracy_score(y_val,valPredict) * 100, 2)) + "%")
print("Val Set Error:", str((round((1 - accuracy_score(y_val,valPredict)) * 100, 2))) + "%")
```

```python
ar = np.arange(len(pokemonList)).tolist()
results = pd.DataFrame({'Ground Truth' : y_val,'predicted' : valPredict, 'Equal': y_val - valPredict == 0})
wrongPer = np.where(results.Equal == False)
wrongDf = results.iloc[wrongPer]
for val in wrongDf.predicted:
  if val in ar:
    ar.remove(val)
print(ar)
for i in ar:
  print(pokemonList[i])
```

```python
#Hyperparameter Search
svc2 = SVC()
param_grid = {'C': [0.1, 1, 10],
              'gamma':['scale', 'auto'],
              'kernel': ['rbf', 'poly', 'sigmoid', "linear"]} 
model = GridSearchCV(svc2, param_grid)
model.fit(X_train, y_train)
```

```python
model.cv_results_
trainPredict2 = model.predict(X_train)
valPredict2 = model.predict(X_val)
#Print Accuracy
print("Training Set Accuracy:", str(round(accuracy_score(y_train,trainPredict2) * 100, 2))+ "%")
print("Test Set Accuracy:", str(round(accuracy_score(y_val,valPredict2) * 100, 2)) + "%")
```
